Insert Functions

```python
import pandas as pd

var = pd.DataFrame({
    "a":[1,2,3,4,5,],
    "b":[3,4,5,6,7]
    })


var.insert(1,"python", [11,2,3,4,4]) #(index of new column, name of it, the list of values)

var
```

Pass limited data

```python
var["python_2"] = var["a"][:3]  #list slicing

var
```

Delete Function

```python
var = pd.DataFrame({
    "a":[1,2,3,4,5,],
    "b":[3,4,5,6,7]
    })

var.pop("b")  #using pop function
var
```

```python
var = pd.DataFrame({
    "a":[1,2,3,4,5,],
    "b":[3,4,5,6,7]
    })

del var["a"]  #using del function

var
```
