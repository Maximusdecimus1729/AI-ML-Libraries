Create CSV File

```python
import pandas as pd

var = pd.DataFrame({
    "a":[1,2,3,4,5],
    "b":[3,4,5,6,6],
    "c":["q","r","y","t","u"]
})

var.to_csv("Test.csv")

```

```python
var = pd.DataFrame({
    "a":[1,2,3,4,5],
    "b":[3,4,5,6,6],
    "c":["q","r","y","t","u"]
})

var.to_csv("Test_2.csv", header = ["alex","bob","click"]) #change column name
```

```python
var = pd.DataFrame({
    "a":[1,2,3,4,5],
    "b":[3,4,5,6,6],
    "c":["q","r","y","t","u"]
})

var.to_csv("Test_1.csv", index = False) #to remove indexes
```

```python
#using Index of column

csv_1 = pd.read_csv("D:\\Panda\\Test_1.csv", usecols = [0,1] )

csv_1
```

Index Col

```python
csv_1 = pd.read_csv("D:\\Panda\\Test_1.csv", index_col = "a")  #column a will act as new index column

csv_1
```

Header parametr

```python
#make the first row as heading

csv_1 = pd.read_csv("D:\\Panda\\Test_1.csv", header = 2)  

csv_1
```

```python
csv_1 = pd.read_csv("D:\\Panda\\Test_1.csv", header = None)  

csv_1
```

name paramter

```python
#change name of headings

csv_1 = pd.read_csv("D:\\Panda\\Test_1.csv", names = ["apple", "cat","Bat"] ) 

csv_1
```

Prefix Parameter

```python
csv_1 = pd.read_csv("D:\\Panda\\Test_1.csv", header = None, prefix = "apple")  

csv_1
```
