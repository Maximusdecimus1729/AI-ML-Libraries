# Merging and Concating

```python
import pandas as pd

sr = pd.DataFrame({
    "s":([1,2,3,4]),
    "r":([2,4,5,6])
})

sr1 = pd.DataFrame({
    "s":([1,2,3,4]),
    "t":([22,24,25,26])
}
)
pd.merge(sr, sr1, on = "s" ) #s is the common
```

```python
sr = pd.DataFrame({
    "s":([1,2,3,4]),
    "r":([2,4,5,6])
})

sr1 = pd.DataFrame({
    "s":([1,2,3,4]),
    "t":([22,24,25,26])
}
)
pd.merge(sr1, sr, on = "s" ) #to change the order of merging
```

```python
sr = pd.DataFrame({
    "s":([1,2,3,4]),
    "r":([2,4,5,6])
})

sr1 = pd.DataFrame({
    "s":([1,2,3,5]),
    "t":([22,24,25,26])
}
)
pd.merge(sr, sr1, on = "s" ) #last element of s is different
```

'How' paramter {inner, left, right, outer}

```python
sr = pd.DataFrame({
    "s":([1,2,3,4]),
    "r":([2,4,5,6])
})

sr1 = pd.DataFrame({
    "s":([1,2,3,5]),
    "t":([22,24,25,26])
}
)
pd.merge(sr, sr1, how = "outer" ) #last element of s is different
```

indicator parameter

```python
sr = pd.DataFrame({
    "s":([1,2,3,4]),
    "r":([2,4,5,6])
})

sr1 = pd.DataFrame({
    "s":([1,2,3,5]),
    "t":([22,24,25,26])
}
)
pd.merge(sr, sr1, how = "outer", indicator = True) #last element of s is different
```

left right index parameter

```python
sr = pd.DataFrame({
    "S":([1,2,3,4]),
    "R":([2,4,5,6])
})

sr1 = pd.DataFrame({
    "S":([1,2,3,5]),
    "R":([22,24,25,26])  #if name of both list is same
}
)
pd.merge(sr, sr1, left_index=True, right_index=True) #to show the complete data
```

Suffix paramter


```python
sr = pd.DataFrame({
    "S":([1,2,3,4]),
    "R":([2,4,5,6])
})

sr1 = pd.DataFrame({
    "S":([1,2,3,5]),
    "R":([22,24,25,26])  #if name of both list is same
}
)
pd.merge(sr, sr1, left_index=True, right_index=True, suffixes = ("_name", "_python")) #change the name of list
```

# Concate

```python
sr1 = pd.Series([1,2,3,4,5,4,5,6,5])
sr2 = pd.Series([12,22,3,34,55])

pd.concat([sr1, sr2] )
```

```python
#concate in data frame : row wise

sr = pd.DataFrame({
    "S":([1,2,3,4]),
    "R":([2,4,5,6])
})

sr1 = pd.DataFrame({
    "S":([1,2,3,5]),
    "R":([22,24,25,26])
}
)
pd.concat([sr, sr1])
```

```python
#concate in data frame : column wise

sr = pd.DataFrame({
    "S":([1,2,3,4]),
    "T":([2,4,5,6])
})

sr1 = pd.DataFrame({
    "U":([1,2,3,5]),
    "V":([22,24,25,26])
}
)
pd.concat([sr, sr1], axis = 1 #no need to have common name
)
```

Join paramter, 'outer' = union, 'inner' = intersection

```python
sr = pd.DataFrame({
    "S":([1,2,3,4]),
    "T":([2,4,5,6])
})

sr1 = pd.DataFrame({
    "U":([1,2]),
    "V":([22,24])
}
)
pd.concat([sr, sr1], axis = 1, join="outer") #no need to have common name
```

```python
sr = pd.DataFrame({
    "S":([1,2,3,4]),
    "T":([2,4,5,6])
})

sr1 = pd.DataFrame({
    "U":([1,2]),
    "V":([22,24])
}
)
pd.concat([sr, sr1], axis = 1, join="inner") #no need to have common name
```

Key paramter


```python
sr = pd.DataFrame({
    "S":([1,2,3,4]),
    "T":([2,4,5,6])
})

sr1 = pd.DataFrame({
    "U":([1,2]),
    "V":([22,24])
}
)
pd.concat([sr, sr1], axis = 1, keys = ["d1", "d2"])
```

concate 1D & 2D dataframe

```python
sr = pd.DataFrame({
    "S":([1,2,3,4])
    
})

sr1 = pd.DataFrame({
    "U":([1,2]),
    "V":([22,24])
}
)
pd.concat([sr, sr1])
```

# Join and Append

```python
var1 = pd.DataFrame({
    "A":[1,2,3,4],
    "B":[11,12,13,14]
})
var2 = pd.DataFrame({
    "C":[10,20],
    "D":[11,22]
})

var1.join(var2)
```

```python
#change index type/number

var1 = pd.DataFrame({
    "A":[1,2,3,4],
    "B":[11,12,13,14],
},
    index = ["a","b","c","d"]
)

var2 = pd.DataFrame({
    "C":[10,20],
    "D":[11,22]
},
    index = ["a","b"]
)

var1.join(var2)

```

```python
var1 = pd.DataFrame({
    "A":[1,2,3,4],
    "B":[11,12,13,14]
})
var2 = pd.DataFrame({
    "C":[10,20],
    "D":[11,22]
})

var2.join(var1, how = "outer") #in reverse order, 'how' to include all values 


```

# Append

```python
var1 = pd.DataFrame({
    "A":[1,2,3,4],
    "B":[11,12,13,14]
})
var2 = pd.DataFrame({
    "C":[10,20],
    "D":[11,22]
})

var1.append(var2)

```
