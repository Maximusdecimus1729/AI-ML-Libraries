Data Structure: Series

```python
import pandas as pd

x = [4,3,5,6,7,8,8,5]

var = pd.Series(x)

print(var)
print(type(var))
print(var[3])
```

```python
#Change Index Type

x = [4,3,5,6]

var = pd.Series(x, index = ["a","b","c","d"], dtype = "float",name = "Python")

print(var)
```

```python
dic = {
    "name":['python','c','c++','java'],
    "populrty":[12,13,14,15],
    "rank":[1,4,6,3]
    }

var1 = pd.Series(dic)

print(var1)
print()
print(dic["name"])
```

```python
series = pd.Series(12, index = [1,2,3,4,5,6,7])

print(series)
```

No broadcast error as in case of Numpy

```python
s1 = pd.Series(12, index = [1,2,3,4,5,6,7])
s2 = pd.Series(12, index = [1,2,3,4])

print(s1+s2) #no error only Nan


```
