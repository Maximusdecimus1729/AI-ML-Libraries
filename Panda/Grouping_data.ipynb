# GroupBy Function

```python
import pandas as pd

var = pd.DataFrame({
    "name":["a","b","b","c","a","c","c","b","a","a"],
    "s_1": [12,13,10,14,12,23,18,15,11,15],
    "s_2": [13,10,14,12,13,10,18,15,11,15]
})

var
```

```python
var_new = var.groupby("name")

var_new #its gruopped but to show it, we need to use loop
```

```python
for x,y in var_new:
    print(x)
    print(y)  #what is y doing here????????????
```
