```python
#INDEXING

import numpy as np

x = np.array([3,4,5,6])

#print index element

print(x[2])

```

```python
#in 2D array

y = np.array([[1,2,3],[4,5,6]])

print(y)
print()
print(y[0,2])

print(y[1,1])
```

```python
#in 3D array

z = np.array([[[1,2],[3,4]]])

print(z)
print()
print(z[0,1,1])


```

```python
#SLICING
import numpy as np
#1D
arr = np.array([1,2,3,4,5])
print(arr)
print()

#to pull out a particular set of element, like 2,3,4 or 3,4,5
print("2 to 5 : ",arr[1:5] )
print("1 to 3 : ",arr[0:3] )
print("2 to end : ",arr[1:] )
print("start to 4 : ",arr[:4] )

#jumping elemnts(start:stop:jump/common diff of index)

print("stop : ",arr[::2])
print("stop : ",arr[3::2])
```

```python
#in 2D
import numpy as np
arr2 = np.array([[4,5,6,5,8,9],[7,4,2,4,5,6],[4,8,9,7,4,2]])
print(arr2)
print()

print(arr2[0,2:5])
print(arr2[0,0:5:2])
print(arr2[2,0:6:3])
print(arr2[2,2::2])

```
