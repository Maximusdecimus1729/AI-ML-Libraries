```python
import numpy as np

#Addition 1D

var = np.array([1,2,3,4])

varAdd = var + 3

print(varAdd)

```

```python
#using functions

arr1 = np.array([1,2,3,4])
arr2 = np.array([5,6,7,8])

add = np.add(arr1, arr2)

print(add)
```

```python
#in 2D array add

arr1 = np.array([[1,2,3], [4,5,6]])
arr2 = np.array([[1,2,3], [4,5,6]])

add  = np.add(arr1, arr2)

print(add)
```

```python
#in 2D array multi
import numpy as np

arr1 = np.array([[1,2,3], [4,5,6]])
arr2 = np.array([[1,2,3], [4,5,6]])

print(arr1)
print()
print(arr2)
print()



multi  = np.multiply(arr1, arr2) #each index is multiplied with each

print(multi)
```

```python
#ARTHMETIC FUNCTIONS
arr1 = np.array([1,2,5,8,5,3,6,2,1,7,8,0])

print("min :",np.min(arr1))
print("max :",np.max(arr1))

#position

print("pos of min :",np.argmin(arr1))
print("pos of max :",np.argmax(arr1))

#sqrt(of individual element)

print("sqrt: ", np.sqrt(arr1))

```

```python
#in 2D

var = np.array([[1,5,8,4] , [2,6,3,8]])

print(np.min(var, axis = 0))

```

```python
#cumsum (cumulutive sum, addding previous terms of the result)

arr1 = np.array([1,2,5,8,5,3,6,2,1,7,8,0])

print("cumsum :", np.cumsum(arr1))
```
