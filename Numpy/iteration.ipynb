```python
#ITERATING

import numpy as np

arr = np.array([1,2,3,4,5])

for  i in arr:
    print(i)
```

```python
#in 2D

arr2 = np.array([[1,3,4],[6,4,3]])

#rowwise iteration
for j in arr2:
    print(j)

#elemntwise
for k in arr2:
    for l in k:
        print(l)


```

```python
#in 3D

arr3 = np.array([[[1,3,4],[6,4,3],[6,7,8]]])

#rowwise iteration
for j in arr3:
    print(j)

#elemntwise
for k in arr3:
    for l in k:
        for m in l:
            print(m)

```

```python
#USING FUNCTIONS{nditer()}

#in 2D

arr2 = np.array([[1,3,4],[6,4,3]])


for i in np.nditer(arr2):
    print(i)

print()

#in 3D

arr3 = np.array([[[1,3,4],[6,4,3],[6,7,8]]])

for j in np.nditer(arr3):
    print(j)

```

```python
#USING FUNCTIONS to get the indexing + datatype of elements too {ndenumerate()}

#in 2D

arr2 = np.array([[1,3,4],[6,4,3]])


for i in np.ndenumerate(arr2):
    print(i)

print()

#in 3D

arr3 = np.array([[[1,3,4],[6,4,3],[6,7,8]]])

for j in np.ndenumerate(arr3):
    print(j)

```
