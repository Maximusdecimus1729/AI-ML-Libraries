```python
#JOIN ARRAY

import numpy as np

var = np.array([1,2,3,4,5])

var1 = np.array([3,5,7,8,5])

join = np.concatenate((var, var1))

print(join)
```

```python
#in 2D
var = np.array([[1,2,3],[5,8,9]])

var1 = np.array([[3,5,7],[4,6,7]])

print(var)
print()
print(var1)
print()

join = np.concatenate((var, var1), axis = 1)

print(join)
print()
join1 = np.concatenate((var, var1), axis = 0)
print(join1)


```

```python
#STACK FUNCTION

var = np.array([1,2,3,4,5])

var1 = np.array([3,5,7,8,5])

join = np.stack((var, var1), axis = 0)
join1 = np.hstack((var, var1)) #row
join2 = np.vstack((var, var1)) #column
join3 = np.dstack((var, var1)) #height

print(join)
print()
print(join1)
print()
print(join2)
print()
print(join3)
print()
```

```python
#SPLITT ARRAY

import numpy as np
#1D

arr = np.array([1,2,3,4,5])

split = np.array_split(arr, 3)

print(split)

print(split[1]) #index of ssplitted array
 
```

```python
#in 2D

arr2 = np.array([[1,2],[3,4],[5,6]])

split1 = np.array_split(arr2, 3)
split2 = np.array_split(arr2, 3, axis = 1)

print(split1)
print()
print(split2)

#for j in np.ndenumerate(split2):
 #   print(j)
```
