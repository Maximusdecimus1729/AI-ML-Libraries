```python
#INSERT
import numpy as np

var = np.array([1,2,4,5])

print(var)

new_var = np.insert(var, 2, 40) #can be also done Append function


#at multiple index
new_var2 = np.insert(var, (2,4), 40) #{array, index, value(only int is accaeptd)}

print(new_var)
print()
print(new_var2)




```

```python
#in 2D

var1 = np.array([[1,2,3],[4,6,6],[7,6,7]])

print(var1)
print()

v1 = np.insert(var1, 2, 16, axis = 1)
v2 = np.insert(var1, 2, 16, axis = 0)
v3 = np.insert(var1, 2, [16,23,56], axis = 1) #new column
v4 = np.insert(var1, 2, [16,23,56], axis = 0) #new row

print(v1)
print()
print(v2)
print()
print(v1)
print()
print(v3)
print()
print(v4)

```
