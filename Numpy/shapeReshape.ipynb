```python
#Reshape

import numpy as np

var2 = np.array([1,2,3,4,5,6])

x = var2.reshape(3,2) #3 = no.of rows, 2 = no. of elements
y = var2.reshape(2,3)
print(x)
print()
print(y)
print()
print(x.ndim)
print()
print(y.ndim)


```

```python
#no broadcast error if the number of rows matched with dusre ke rows,
import numpy as np

x = np.array([[1],[2]]) 
print(x.shape)

y = np.array([[1,2,3],[1,2,3]])
print(y.shape)

add = np.add(x,y)

print(add)

```
