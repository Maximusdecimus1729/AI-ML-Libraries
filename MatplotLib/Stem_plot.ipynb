```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [3,5,2,4,1,5]

plt.stem(x,y)
plt.title("Stem Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

line format and marker

```python
x = [1,2,3,4,5,6]
y = [3,5,2,4,1,5]

plt.title("Stem Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.stem(x,y, linefmt = ":", markerfmt = "r+") # red and + symbol at pkace of dot 

plt.show()
```

Bottom paramter


```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [3,5,2,4,1,5]

plt.title("Stem Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")


plt.stem(x,y, bottom = 1, basefmt = "y", label = "python") 

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [3,5,2,4,1,5]

plt.title("Stem Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")


plt.stem(x,y, bottom = 1, basefmt = "y,", label = "python", orientation = "horizontal") 

plt.legend()
plt.show()
```
