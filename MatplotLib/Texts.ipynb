Different formats of Text


```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [3,12,4,5,6,8]

plt.plot(x,y)
plt.title("python uses", fontsize = 20)

plt.xlabel("Days", fontsize = 15)
plt.ylabel("Percent", fontsize = 15)

#Text att Random location

plt.text(2,12, "User1", style = "italic")
plt.text(3,4, "User2", bbox = {"facecolor" : "red"})

#annotations

plt.annotate("Gaurav", xy =(5,6), xytext = (4,10),
        arrowprops = dict(facecolor = "black", shrink = 1000))


plt.show()
```
