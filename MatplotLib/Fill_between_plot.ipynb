```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,3,5,6,7]

plt.title("Color Between")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.fill_between(x,y)

plt.plot(x,y, color = "r")

plt.show()
```

Fill in an Area


```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,3,5,6,7]

plt.title("Color Between")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.fill_between(x=[2,3.5],y1 = 3, y2 = 5)

plt.plot(x,y, color = "r")

plt.show()
```

Fill in a particular range

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([2,3,5,6,7]) #because this function works only with Array not list 


plt.title("Color Between")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.fill_between(x,y, where = (x>=2) & (x<=3.5), label = "coloured")

plt.plot(x,y, color = "r")
plt.legend(loc = 2)
plt.show()
```
