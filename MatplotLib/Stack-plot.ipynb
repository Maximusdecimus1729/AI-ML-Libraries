```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
area = [2,1,5,3,4]
area1 = [2,2,7,4,6]
area2 = [4,3,9,3,4]

l = ["area","area1","area2"]

plt.stackplot(x, area, area1, area2, labels = l)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
area = [2,1,5,3,4]
area1 = [2,2,7,4,6]
area2 = [4,3,9,3,4]

l = ["area","area1","area2"]

plt.stackplot(x, area, area1, area2, labels = l, baseline = "sym")


plt.title("Python")
plt.xlabel("x-axis")
plt.legend()
plt.show()
```
