Grid parametr

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,4,3,7,8,9]

plt.step(x,y)

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-label")

plt.grid()

plt.show()
```

Marker(vertex), marker color, marker size

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,4,3,7,8,9]

plt.step(x,y, color = "r", marker = "o", ms = 5.5, mfc = "g")

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-label")

plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,4,3,7,8,9]

plt.step(x,y, color = "r", marker = "o", ms = 5.5, mfc = "g", label = "language")

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.legend(loc = 2)

plt.grid()

plt.show()
```
