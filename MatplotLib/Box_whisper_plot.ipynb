Label, color, mean

```python
import matplotlib.pyplot as plt

x  = [10,30,50,20,30,40,10] #the order of elements does matter
x1 = [10,20,30,40,50,60,70, 120]


plt.boxplot(x1, widths =0.1, tick_labels = ["Python"], patch_artist = True, showmeans = True,  ) #patch artist for color
plt.title("Box Plot")
plt.show()

```

outlier, box color, cap color, wisk color

```python
import matplotlib.pyplot as plt

x  = [10,30,50,20,30,40,10] #the order of elements does matter
x1 = [10,20,30,40,50,60,70, 120]


plt.boxplot(x1, sym = "g*", boxprops = dict(color = "r"), capprops = dict(color = "r"), whiskerprops = dict(color = "r") )
plt.show()

```

Multiple box

```python
import matplotlib.pyplot as plt

x  = [10,30,50,20,30,40,10] #the order of elements does matter
x1 = [10,20,30,40,50,60,70, 120]


plt.boxplot([x, x1], tick_labels = ["python", "c++"])
plt.show()

```
