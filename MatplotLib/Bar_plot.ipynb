```python
import matplotlib.pyplot as plt

x = ["pyhton","c","java","c++"]
y = [85,77,56,89]

plt.xlabel("Langauages")
plt.ylabel("Percen.")
plt.title("Uses", fontsize = 15)

#different paramaters

plt.bar(x,y, width = 0.5, color = "b", align = "center",
edgecolor  ="r", linewidth = 5, linestyle = ":", alpha = 0.4,
label = "Popularity" )

plt.legend() #for legend parametr
plt.show()
```

## Two bar graphs vertical

```python
import numpy as np

x = ["pyhton","c","java","c++"]
y = [85,77,56,89]
z = [20,30,40,50]

width = 0.4
p = np.arange(len(x))  #to make an array from the list
p1 = [j + width for j in p] #tpo displace the the second bar from the one horizontely

plt.xlabel("Langauages")
plt.ylabel("Percen.")
plt.title("Uses", fontsize = 15)

plt.bar(p,y, width = 0.4, label = "popularity")
plt.bar(p1,z, width = 0.4, label = "popularity_2")

plt.xticks(p + width/2, x) #to get the x axis label at the center
plt.legend()
plt.show()
```

## two bar graphs horizontal

```python
import numpy as np

x = ["pyhton","c","java","c++"]
y = [85,77,56,89]
z = [20,30,40,50]

width = 0.4
p = np.arange(len(x))  #to make an array from the list
p1 = [j + width for j in p] #tpo displace the the second bar from the one horizontely

plt.xlabel("Langauages")
plt.ylabel("Percen.")
plt.title("Uses", fontsize = 15)

plt.barh(p,y, width, label = "popularity")
plt.barh(p1,z, width, label = "popularity_2")

plt.xticks(p + width/2, x)
plt.legend()
plt.show()
```
