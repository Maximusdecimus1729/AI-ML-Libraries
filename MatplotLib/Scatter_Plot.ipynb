```python
import matplotlib.pyplot as plt

Day = [1,2,3,4,5,6,6,7]
No = [2,3,5,7,9,8,6,2]

clr = ["r","g","b","y","g","r","r","g"]
sizes = [800,120,320,350,120,760,120,990]

plt.scatter(Day,No, s = sizes, c = clr, alpha = 0.2, #marker = "*",
            edgecolor = "r" ) #marker to give the shape

plt.title("Scatter Plot", fontsize = 20)
plt.xlabel("Day", fontsize = 10)
plt.ylabel("No", fontsize = 10)

plt.show()
```

# color bar

```python
import matplotlib.pyplot as plt

Day = [1,2,3,4,5,6,6,7]
No = [2,3,5,7,9,8,6,2]

clr = ["r","g","b","y","g","r","r","g"]
sizes = [800,120,320,350,120,760,120,990]

plt.scatter(Day,No, s = sizes, c = clr, cmap = "PuRd")
            
#plt.colorbar()
t = plt.colorbar()
t.set_label("Match Karlo")

plt.title("Scatter Plot", fontsize = 20)
plt.xlabel("Day", fontsize = 10)
plt.ylabel("No", fontsize = 10)

plt.show()
```

# Double Scatter

```python
import matplotlib.pyplot as plt

Day = [1,2,3,4,5,6,6,7]
No = [2,3,5,7,9,8,6,2]
No2 = [1,7,2,4,8,9,7,9]

clr = ["r","g","b","y","g","r","r","g"]
clr2 = ["r","r","g","g","b","y","r","g"]
sizes = [800,120,320,350,120,760,120,990]

plt.scatter(Day,No, s = sizes, c = clr)
plt.scatter(Day,No2, s = sizes, c = clr2)
            
plt.title("Scatter Plot", fontsize = 20)
plt.xlabel("Day", fontsize = 10)
plt.ylabel("No", fontsize = 10)

plt.show()
```
