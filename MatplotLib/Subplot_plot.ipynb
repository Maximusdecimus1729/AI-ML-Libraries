More than two graphs in one plot

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,3,5,6,3,7]
plt.subplot(2,2,1) #Subplot here
plt.plot(x,y)

plt.subplot(2,2,2) #Subplot here
plt.pie([1], colors = "r")


x1 = ['a','b','c','d','e']
y1 = [10,20,30,40,50]
plt.subplot(2,2,4) #Subplot here
plt.bar(x1,y1)


x2 = [10,20,30,40]
plt.subplot(2,2,3) #Subplot here
plt.pie(x2)


plt.show()
```
