```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ["c","c++","Java","Python"]

plt.pie(x, labels = y)
plt.show()
```

Explode paramter

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ["c","c++","Java","Python"]
ex = [0.4, 0.0, 0.0, 0.0]

plt.pie(x, labels = y, explode = ex)
plt.show()
```

Percentage (Autopct paramter)

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ["c","c++","Java","Python"]
ex = [0.4, 0.0, 0.0, 0.0]

plt.pie(x, labels = y, explode = ex, autopct = "%0.1f%%") #0.2 for 2 decimal  places
plt.show()
```

Radius, Shadow, Label Distance, Rotation, FontSize, Wedgeprops


```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ["c","c++","Java","Python"]
ex = [0.4, 0.0, 0.0, 0.0]

plt.pie(x, labels = y, explode = ex, autopct = "%0.1f%%", shadow = "True", radius = 1.3, labeldistance = 1.23, startangle = 90, textprops = {"fontsize": 15}, rotatelabels = True)
 
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ["c","c++","Java","Python"]
ex = [0.4, 0.0, 0.0, 0.0]

plt.pie(x, labels = y, explode = ex, autopct = "%0.1f%%")
plt.title("Pie Chart")
plt.legend(loc = 2)

plt.show()
```

two pie charts / Donut pie chart


```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
x1 = [40,24,56,70]
y = ["c","c++","Java","Python"]

plt.pie(x, labels = y)
plt.pie(x1, radius = 0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ["c","c++","Java","Python"]

plt.pie(x, labels = y, radius = 1.5)

plt.pie([1], colors = "w") #using Dot Pie chart
plt.show()
```

```python
#Another method for same

import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ["c","c++","Java","Python"]

plt.pie(x, labels = y, radius = 1.5)

#new stuff, too create a circle
cr = plt.Circle(xy = (0,0), radius = 1, facecolor = "w")
plt.gca().add_artist(cr)

plt.show()
```
