To change the limits of inputs on either axis.

```python
import matplotlib.pyplot as gc

x = [1,2,3,4,5]
y = [3,1,5,2,4]

gc.plot(x,y)

gc.show()
```

xticks and yticks

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]

gc.plot(x,y)

gc.xticks(x) #jo given data hoga wahi axis pr show hoga
gc.yticks(y)

gc.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]

gc.plot(x,y)
gc.xticks(x, labels = ["pyhton", "java", "c", "c++", "html"])
gc.yticks(y)

gc.show()
```

Limits(xlim, ylim)

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]

gc.plot(x,y)

gc.xlim(1,10)
gc.ylim(1,10)

gc.show()



```
